# Teste de uma amostra para médias

Utiliza-se um teste de uma amostra para testar uma afirmação sobre uma única média populacional. Extrai-se uma amostra de $n$ observações e calcula-se a média amostral. Compara-se então o desvio entre o valor alegado e esta média amostral com avariabilidade da distribuição amostral baseada na afirmação. Grandes desvios sugerem que a afirmação é falsa; pequenos desvios corroboram a afirmação.

Suponha-se que queiramos avaliar a afirmação de um fabricante, de que seus pneus radiais suportam uma quilometragem de 40.000 milhas. A hipótese nula é 

$$H_0 : \mu = 40.000 \text{ milhas}$$

Para ilustrar, consideremos as três hipóteses alternativas possíveis

$$H_1 : \mu \neq 40.000 \text{ milhas}$$
$$H_1 : \mu > 40.000 \text{ milhas}$$
$$H_1 : \mu < 40.000 \text{ milhas}$$

Escolhido o nível de significância, podem coletar os dados amostrais e calcular a estatística de teste.

$$\text{Estatística de Teste} = \frac{\text{média amostral} - \text{média alegada}}{\text{desvio padrão da distribuição amostral}}$$

Se o desvio padrão da população é conhecido, a estatística é

$$z_{teste} = \frac{\bar{x} - \mu_0}{\sigma_x / \sqrt{n}}$$

Se $\sigma_x$ é desconhecido, a estatística é

$$t_{teste} = \frac{\bar{x} - \mu_0}{s_x / \sqrt{n}}$$

---

Testando a hipótese de que a média é menor que 40.000 milhas com $\alpha = 0.05$. Tirando uma amostra de $n = 49$, com média amostral 38.000 e sabendo-se que a população tem desvio padrão de 3.500 milhas:

$$H_1 : \mu < 40.000 \text{ milhas}$$

Calculando a estatística de teste:

$$z_{teste} = \frac{\bar{x} - \mu_0}{\sigma_x / \sqrt{n}} = \frac{38.000 - 40.000}{3.500 \sqrt{49}} = -4,0$$

Como a estatística de teste (-4) excede o valor crítico adotado (-1,65), podemos rejeitar a hipótese nula $H_0$ concluindo que a vida média dos pneus é infeior a 40.000 milhas (Não precisamos, portanto, testar $\mu > 40.000$). Note-se que este resultado *não* garante que a afirmação seja falsa. Poderemos estar cometendo um erro Tipo I. Assim é que, sempre que se rejeita $H_0$, isto significa que a evidência amostral *sugere* que $H_0$ é falsa; não há maneira de se estar *absolutamente certo* se a afirmação é verdadeira ou não, a menos que se proceda a um censo completo.

Como decidimos que a média verdadeira não é 40.000 milhas, uma pergunta lógica é: qual é a verdadeira vida média dos pneus? Uma forma de responder a esta pergunta é usar os dados amostrais para estimar a média. Usando um nível de confiança de $(1-\alpha)$, ou seja, 95%, temos

$$\bar{x} \pm 1,96 \frac{\sigma_x}{\sqrt{n}} = 38.000 \pm 1,96(500) = 38.000 \pm 980$$

---

Agora, vamos testar $H_1 : \mu \neq 40.000$

Já temos a estatística de teste calculada como -4 o que excede o valor crítico inferior de referência (1,96). Portanto concluímos pela rejeição de $H_0$.

---

### $\sigma_x$ desconhecido

Quando não se conhece o desvio padrão da população, deve-se estimá-lo a partir dos dados amostrais usando o desvio padrão amostral. A distribuição $t$ é a mais adequada nesse caso. Na prática, entretanto, só se exige o uso da distribuição $t$ quando o tamanho da amostra é igual ou inferior a 30, pois, para maiores valores do tamanho da amosltra, os valores de $t$ e $z$ são aproximadamente os mesmos, embora a estatística de teste seja sempre referida como $t$ quando se usa $s_x$, independentemente do tamanho da amostra.

#### Exemplo 1 (grandes amostras)

Verifiquemos as mesmas hipóteses alternativas:

$$H_1 : \mu \neq 40.000 \text{ milhas}$$
$$H_1 : \mu > 40.000 \text{ milhas}$$
$$H_1 : \mu < 40.000 \text{ milhas}$$


Suponhamos uma amostra com 36 observações com média 41.200 e desvio padrão amostral 3000. Usemos $\alpha = 0.05$. Ora, como $n$ é maior que 30, podemos usar o valor de $z$ de uma tabela normal para aproximar o valor de $t$. Logo, o valor crítico é +1,65 ou -1,65 para um teste unilateral e $\pm 1,96$ para um teste bilateral.

Para qualquer teste a estatística é

$$z_{teste} = \frac{\bar{x} - \mu_0}{\sigma_x / \sqrt{n}} = \frac{41.200 - 40.000}{3.000 \sqrt{36}} = +2,40$$

As três hipóteses alternativas são testadas. Note-se que a finalidade dos três testes é ilustrar as várias possibilidades; na prática só seria utilizada uma alternativa. Concluímos pela **rejeição** de $H_0$.

#### Exemplo 2 (pequenas amostras)

Suponhamos que o tamanho da amostra é 25. Nesse caso, buscamos o valor de $t$ numa tabela da distribuição. Para essas amostra, os resultados foram:

$$\bar{x} = 41.100 \text{ milhas} \quad \quad s_x = 2.750 \text{ milhas}$$

Usando $\alpha = 0,05$ e $25 - 1 = 24$ graus de liberdade, o valor crítico de $t$ é +1,71 ou -1,71 para um teste unilateral e $\pm 2,07$ para um teste bilateral. A estatística de teste é

$$t_{teste} = \frac{\bar{x} - \mu_0}{s_x / \sqrt{n}} = \frac{41.100 - 40.000}{2.750 \sqrt{25}} = +2,0$$

Para um teste bicaudal, **não é possível rejeitar a hipótese nula** visto que 2,0 não excede o valor crítico 2,07. Nesse sentido, concluímos pela veracidade da hipótese nula de que a média é igual a 40.000 milhas. 

Entretanto, se efetuarmos o teste unicaudal superior, encontramos uma estatística de teste maior do que o valor crítico (+1,71).

Em suma, podemos afirmar com 95% de confiança que a vida média dos pneus é superior a 40.000 milhas mas não podemos afirmar com a mesma confiança que a vida média dos pneus é diferente de 40.000.

As figuras abaixo ilustram os três teste de hipótese:

```julia
using StatsPlots, Distributions, KernelDensity, LaTeXStrings
```

```julia
function npdensity(z, corte; tipo="bilateral")
    n = size(z,2)
    y = kde(z)
    desv = std(z)
    m = mean(z)
    
    if tipo == "bilateral"
        Plots.plot(range(-(desv*3), stop=-(desv*2.07), length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:grey), label="Rejeição", size=(400,300))
        Plots.plot!(range(-(desv*2.07), stop=(desv*2.07), length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:white), label="Aceitação")
        Plots.plot!(range(desv*2.07, stop=desv*3, length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:grey), label="Rejeição")
    end
    if tipo == "superior"
        Plots.plot(range(-(desv*3), stop=(desv*1.71), length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:white), label="Aceitação", size=(400,300))
        Plots.plot!(range((desv*1.71), stop=(desv*3), length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:grey), label="Rejeição")
    end
    if tipo == "inferior"
        Plots.plot(range(-(desv*3), stop=-(desv*1.71), length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:grey), label="Rejeição", size=(400,300))
        Plots.plot!(range(-(desv*1.71), stop=(desv*3), length=100),z->pdf(y,z), color=:blue, 
                    fill=(0,0.5,:white), label="Aceitação")
    end
    vline!([corte], label="Estatística teste", linestyle=:dash, color=:black)
end
```

```julia
vetor = rand(TDist(24), 10000000)
npdensity(vetor, 2, tipo="inferior")
title!("Teste unicaudal inferior")
```

```julia
vetor = rand(TDist(24), 10000000)
npdensity(vetor, 2, tipo="superior")
title!("Teste unicaudal superior")
```

```julia
vetor = rand(TDist(24), 10000000)
npdensity(vetor, 2, tipo="bilateral")
title!("Teste bilateral")
```

# Teste de duas amostras para médias

Os teste de duas amostras são usados para decidir se as médias de duas populações são iguais. Exigem-se duas amostras independentes, uma de cada população (isto é, de grupos diferentes).

Os testes de duas amostras são frequentemente usados para comparar dois métodos de ensino, duas marcas, duas cidades, dois distritos escolares, e outros casos análogos.

A hipótese nula pode ser a de que as duas populçações têm médias iguais:

$$H_0 : \mu_1 = \mu_2$$

As hipóteses alternativas podem ser:

$$H_1 : \mu_1 \neq \mu_2$$
$$H_1 : \mu_1 > \mu_2$$
$$H_1 : \mu_1 < \mu_2$$

A diferença entre as médias é dividida pelo desvio padrão de uma distribuição amostral. Calcula-se primeiro o desvio padrão, supondo $H_0$ verdadeira. Em tal caso, as duas amostras podem ser consideradas como provenientes da *mesma* população, e mediante combinação (*pooling*) das variâncias das duas populações (ou das duas amostras, se as variâncias da população são desconhecidas), pode-se determinar a variância global da população. Quando $\sigma_1$ e $\sigma_2$ são conhecidos a estatística é a seguinte:

$$z_{teste} = \frac{\overline{x}_1 - \overline{x}_2}{\sqrt{\frac{{\sigma_1}^2}{n_1} + \frac{{\sigma_2}^2}{n_2}}}$$

Com o desvio padrão desconhecido:

$$t_{teste} = \frac{\overline{x}_1 - \overline{x}_2}{\sqrt{\frac{{s_1}^2}{n_1} + \frac{{s_2}^2}{n_2}}}$$

Quando os tamanhos das duas amostras não são iguais e sua soma é menor que 30, a fórmula da estatística de teste é

$$t_{teste} = \frac{\overline{x}_1 - \overline{x}_2}{ \sqrt{ \left[ \frac{(n_1 - 1){s_1}^2 + (n_2 - 1){s_2}^2}{n_1 + n_2 - 2} \right] \left( \frac{1}{n_1} + \frac{1}{n_2} \right) }} $$

O valor de $t$ quando $H_0$ é verdadeira tem distribuição $t$ com $n_1 + n_2 - 2$ graus de liberdade, desde que possa admitir que ambas as populações sejam aproximadamente normais. 

#### Exemplo 1

$\sigma_x$ conhecido. $\alpha = 0,05$.

\begin{align}
H_0 : \mu_1 = \mu_2 \quad & \quad H_1 : \mu_1 > \mu_2 \\
\overline{x_1} = 20 \quad & \quad \overline{x_2} = 18 \\
\sigma_1 = \sigma_2 = 3 \quad & \\
n_1 = n_2 = 36 \quad & \\
\end{align}

Calculando a estatística de teste:

$$z_{teste} = \frac{\overline{x}_1 - \overline{x}_2}{\sqrt{\frac{{\sigma_1}^2}{n_1} + \frac{{\sigma_2}^2}{n_2}}} = \frac{20 - 18}{\sqrt{\frac{9}{36} + \frac{9}{36}}} = \frac{2}{\sqrt{0,5}} = \frac{2}{0,7} = 2,9$$

Como a estatística teste é maior que 1,65, rejeitaremos $H_0$ e concluiremos que a média da população 1 é maior que a da população 2.

#### Exemplo 2

$\sigma_x$ desconhecido.

Testar a hipótese:

$$H_0 : \mu_1 = \mu_2$$
$$H_1 : \mu_1 \neq \mu_2$$

Use $\alpha = 0,05$.

$$\overline{x_1} = 5,4 \quad \quad \overline{x_2} = 5,0$$
$$s_{x_1} = 1,1 \quad \quad s_{x_2} = 1,2$$

---

(a) Use $n1 = n2 = 36$

Calculando a estatística de teste:

$$t_{teste} = \frac{\overline{x}_1 - \overline{x}_2}{\sqrt{\frac{{s_{x_1}}^2}{n_1} + \frac{{s_{x_2}}^2}{n_2}}} = \frac{5,4 - 5,0}{\sqrt{\frac{1,21}{36} + \frac{1,44}{36}}} = \frac{0,4}{\sqrt{0,2 + 0,24}} = \frac{0,4}{0,663} = 0,6$$

Como $n_1$ + $n_2 > 30$, $z \approx t$. Como a estatística de teste está na região de aceitação, a hipótese nula não pode ser rejeitada. Concluímos então que a diferença entre as duas médias amostrais é provavelmente o resultado da variação casual devida à amostragem aleatória.

---

(b) Use $n1 = 15$, $n2 = 36$

Calculando a estatística de teste:


\begin{align}
t_{teste}   & = \frac{\overline{x}_1 - \overline{x}_2}{ \sqrt{ \left[ \frac{(n_1 - 1){s_1}^2 + (n_2 - 1){s_2}^2}{n_1 + n_2 - 2} \right] \left( \frac{1}{n_1} + \frac{1}{n_2} \right) }} \\
            & = \frac{5,4 - 5,0}{ \sqrt{ \left[ \frac{(14){1,1}^2 + (2){1,2}^2}{15 + 3 - 2} \right] \left( \frac{1}{15} + \frac{1}{3} \right) }} \\
            & = \frac{0,4}{0,704} \\
            & = 0,568
\end{align}

Vamos testar se a média de renda da população é igual a 1500 reais. Primeiro vamos implementar o teste $t$ "na mão".

```julia
# Vamos encapsular em uma função apenas para testar o tempo de execução
function t_test(data::Array{Float64}, mu::Int64)
    xBar, n = mean(data), length(data)
    sigma = std(data)
    mu0A = mu
    testStatistic = ( xBar - mu0A ) / ( sigma/sqrt(n) )
    pVal = 2*ccdf(Normal(), abs(testStatistic))
    println("Results for mu0 = ", mu0A,":")
    println("Manually calculated test statistic: ", testStatistic)
    println("Manually calculated p-value: ", pVal,"\n")
end
```

```julia
@time t_test([i for i in skipmissing(bd.V4718)], 1500)
```

Agora, vamos testar se duas amostras possuem a mesma média de renda, a saber, homens e mulheres. Primeiro vamos implementar "na mão".

```julia
function t_test2(data1::Array{Float64}, data2::Array{Float64})
    xBar1, s1, n1 = mean(data1), std(data1), length(data1)
    xBar2, s2, n2 = mean(data2), std(data2), length(data2)
    delta0 = 0
    sP = sqrt( ( (n1-1)*s1^2 + (n2-1)*s2^2 ) / (n1 + n2 - 2) )
    testStatistic = ( xBar1-xBar2 - delta0 ) / ( sP * sqrt( 1/n1 + 1/n2) )
    pVal = 2*ccdf(TDist(n1+n2 -2), abs(testStatistic))
    println("Manually calculated test statistic: ", testStatistic)
    println("Manually calculated p-value: ", pVal, "\n")
end
```

```julia
@time t_test2([i for i in skipmissing(bd[bd.V0302 .== "Masculino",:V4718])],
            [i for i in skipmissing(bd[bd.V0302 .== "Feminino",:V4718])])
```

Agora, vamos realizar o teste de duas médias assumindo variância diferente:

```julia
function test_t_unequal(data1::Array{Float64}, data2::Array{Float64})
    xBar1, s1, n1 = mean(data1), std(data1), length(data1)
    xBar2, s2, n2 = mean(data2), std(data2), length(data2)
    delta0 = 0
    v = ( s1^2/n1 + s2^2/n2 )^2 / ( (s1^2/n1)^2/(n1-1) + (s2^2/n2)^2/(n2-1) )
    testStatistic = ( xBar1-xBar2 - delta0 ) / sqrt( s1^2/n1 + s2^2/n2 )
    pVal = 2*ccdf(TDist(v), abs(testStatistic))
    println("Manually calculated degrees of freedom, v: ", v)
    println("Manually calculated test statistic: ", testStatistic)
    println("Manually calculated p-value: ", pVal, "\n")
end
```

```julia
@time test_t_unequal([i for i in skipmissing(bd[bd.V0302 .== "Masculino",:V4718])], 
                   [i for i in skipmissing(bd[bd.V0302 .== "Feminino",:V4718])])
```
