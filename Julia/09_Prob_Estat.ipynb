# Análise da Variância

>Aplica-se a *análise de variância* para determinar se as médias de duas ou mais populações são iguais.

#### Suposições:

1. As amostras devem ser aleatórioas e independentes.
2. As amostras devem ser extraídas de populações normais.
3. As populações devem ter variâncias iguais (isto é, ${\sigma_1}^2 = {\sigma_2}^2 = \dots = {\sigma_k}^2$).

### Revisão do processo de cálculo de uma variância amostral

A análise de variância, como o nome indica, envolve o cálculo de variâncias. A variância de uma amostra é a média dos quadrados dos desvios em relação à media do grupo. Simbolicamente

$$\sigma^2 = \frac{\sum (x_i - \overline{x})}{n-1}$$

## Fundamentos Lógicos da Análise de Variância

Embora possa parecer estranho, o exame das *variâncias* pode revelar se as *médias* populacionais são ou não são iguais. A análise de variância utiliza dois processos um tanto diferentes para estimar variâncias (iguais) das populações. Se as duas estimativas são aproximadamente iguais, isto tende a confirmar $H_0$; se uma das estimativas é muito maior que a outra, isto tende a confirmar $H_1$. Vejamos como realizar as duas estimativas.

Se a hipótese nula é verdadeira, então todas as amostras provêm de populações com médias iguais. E como se supõe que todas as populações sejam normais e tenham variâncias iguais, quando $H_0$ é verdadeira isto é conceitualmetne idêntico a uma situação em que todas as amostras tenham sido extraídas de uma *única* população. Tal situação é ilustrada no primeiro gráfico abaixo. Se $H_0$ é falsa, então as amostras provêm de populações com médias diferentes (cf. segundo gráfico abaixo). Note-se, entretanto, que, mesmo nesse caso, devemos ainda supor que as populações sejam normais e tenham variâncias iguais.

```julia
using Distributions, StatsKit, StatsPlots
```

```julia
plot(Normal(0,1), title="Quatro amostras com médias iguais", legend=false, alpha=.6, width=2)
plot!(Normal(0,1), alpha=.6, width=2)
plot!(Normal(0,1), alpha=.6, width=2)
plot!(Normal(0,1), alpha=.6, width=2)
```

```julia
plot(Normal(6,1), title="Quatro amostras com médias diferentes", legend=false, alpha=.8, width=2)
plot!(Normal(1,1), legend=false, alpha=.8, width=2)
plot!(Normal(-1,1), legend=false, alpha=.8, width=2)
plot!(Normal(14,1), legend=false, alpha=.8, width=2)
```
