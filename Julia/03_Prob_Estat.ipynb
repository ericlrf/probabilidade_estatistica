## Variáveis aleatórias

> Uma variável aleatória é uma função com valores numéricos, cujo valor é determinado por fatores de chance.
>
> Uma variável aleatória é considerada *discreta* se toma valores que podem ser *contados*.
>
> Uma variável aleatória é considerada *contínua* quando pode tomar *qualquer* valor de determinado intervalo.

## Valor esperado de uma variável aleatória

Se uma variável aleatória x toma os valores $x_1$, $x_2$, $x_3$, ..., $x_n$, com as probabilidades correspondentes $p_1$, $p_2$, $p_3$, ..., $p_n$, então seu *valor esperado*, $E(x)$ é 

$$p_1x_1 + p_2x_2 + p_3x_3 + ... + p_nx_n$$

Assim, o *valor esperado* de um experimento é uma média, e pode ser calculado como:

$$E(x) = \sum_{i=1}^{n}p_ix_i$$

#### Exemplo

Suponha que uma loja tenha compilado os seguintes dados sobre vendas de refrigeradores:

| Número vendido $(x_i)$ | Frequência relativa $P(x)$ |
|------------------------|----------------------------|
| 0 | 0,20 |
| 1 | 0,30 |
| 2 | 0,30 |
| 3 | 0,15 | 
| 4 | 0,05 |
| TOTAL  | 1,00 |

Desse modo, 

$$E(x) = 0,20(0) + 0,30(1) + 0,30(2) + 0,15(3) + 0,05(4) = 1,55$$

Como a firma não pode vender 1,55 refrigeradores em nenhum dia a pergunta óbvia é: Como interpretar aquele valor? O valor esperado é uma média a longo prazo.

## Soma de Variáveis Aleatórias

Há situações em que precisamos considerar uma variável aleatória em que ela mesma é a soma de duas ou mais variáveis aleatórias. Em tais casos, devemos ser capazes de determinar a média e o desvio padrão da variável aleatória resultante. Sendo a média e o desvio padrão das variáveis originais conhecidos, 

$$\mu_{x + y} = \mu_x + \mu_y  $$ 

$$\sigma_{x + y} = \sqrt{{\sigma_x}^2 + {\sigma_y}^2}$$

> A média de duas ou mais variáveis aleatórias ´igual à soma das médias das variáveis aleatórias.
>
> A variância de duas ou mais variáveis aleatórias é igual à soma das variâncias dessas variáveis aleatórias.

# Distribuições de Probabilidades

Uma *distribuição de probabilidades* é uma distribuição de frequências para os resultados de um espaço amostral (isto é, para os resultados de uma variável aleatória). As frequências são *relativas*, ou *probabilidades*. Assim, as probabilidades indicam a percentagem de vezes que, em grande número de observações, podemos esperar a ocorrência dos vários resultados de uma variável aleatória.

Pode-se achar, sem muito esforço, a probabilidade de um dado resultado, ou de um grupo de resultados. Do ponto de vista prático, em geral, não é necessário calcular as probabilidades individuais para obter uma distribuição de probabilidades. Existem tabelas e fórmulas para isso. Consequentemente, o problema não é "como se *deduzem* os valores?", mas sim, "como se *usam* as distribuições de probabilidades para resolver problemas?"

Fora o fato de que as distribuições de probabilidades proporcionam um método simples para a determinação de certas probabilidades, os tipos de distribuição podem ser considerados como *modelos* para descrever situações que envolvem resultados gerados pela chance.

# Distribuições Descontínuas

As distribuições descontínuas de probabilidades envolvem variáveis aleatórias relativas a dados que podem ser contados. Estudaremos duas importantíssimas distribuições descontínuas, a binomial e a de Poisson.

## A Distribuição Binomial

Usa-se o termo binomial para designar situações em que os resultados de uma variável aleatória podem ser agrupados em duas classes ou categorias. O dados são, pois, *nominais*. As categorias devem ser mutuamente excludentes, de modo a deixar perfeitamente claro a qual categoria pertence determinada observação; e as classes devem ser coletivamente exaustivas, de forma que nenhum outro resultado fora delas é possível.

A utilização da distribuição binomial exige certas hipótese:

1. Há $n$ observações ou provas idênticas;
2. Cada prova tem dois resultados possíveis, um chamado "sucesso" e o outro "falha" (ou "fracasso");
3. As probabilidades $p$ de sucesso e $1 - p$ de falha permanecem constantes em todas as provas;
4. Os resultados das provas são independentes uns dos outros.

Há dois métodos para obtenção das probabilidades para uma variável aleatória distribuída binomialmente. Um deles consiste em utilizar a fórmula binomial, o outro é a utilização das tabelas de probabilidades binomiais.

## A fórmula binomial

$$P(x) = {n \choose x} P(sucesso)^x P(falha)^{n-x}$$

onde ${n \choose x}$ é o número de maneiras de obter $x$ sucessos e $n-x$ falhas em $n$ provas.

## Tabelas binomiais

Há dois tipos de tabelas binomiais, Um dá as probabilidades de resultados *individuais* de uma v.a., e o outro dá as probabilidades de um *conjunto* de resultados. 

# Características da Distribuição Binomial

A *média* de uma distribuição binomial é a média a longo prazo, ou o *valor esperado*, de uma v.a. binomial. O *desvio padrão* de uma distribuição binomial indica até que ponto os valores amostrais tendem a se afastar da média da distribuição. No caso da binomial, tanto a média como o desvio padrão podem ser expressos em termos do *número* ou da *percentagem* de sucessos.

Número de sucessos - Média = $np$. Desvio padrão = $\sqrt{np(1-p}$

Percentagem de sucessos - Média = $p$. Desvio padrão = $\sqrt{p(1-p)/n}$

```julia
# Implementando uma função para calcular a probabilidade binomial
function prob_binom(x, n, p; rounded=true)
    res = binomial(n, x) * (p^x * (1-p)^(n-x))
    if rounded
        return round(res, digits=4)
    else
        return res
    end
end
```

# A distribuição de Poisson

A distribuição de Poisson é útil para descrever as probabilidades do número de ocorrências num campo ou intervalo contínuo (em geral tempo ou espaço). Alguns exemplos: defeitos por centímetro quadrado, acidentes por dia, clientes por hora, chamadas telefônicas por minuto, etc. Note-se que a unidade de medida (tempo, área) é contínua mas a variável aleatória (número de ocorrências), é discreta. Além disso, *as falhas não são contáveis*. Não é possível contar os acidentes que não ocorreram.

A utilização da distribuição de Poisson baseia-se nas seguintes hipóteses:

1. A probabilidade de uma ocorrência é a mesma em todo o campo de observações;
2. A probabilidade de mais de uma ocorrência num único ponto é aproximadamente zero;
3. O número de ocorrências em qualquer intervalo é independente do número de ocorrências em outros intervalos.

Abaixo, vejamos uma distribuição de Poisson típica.

```julia
?Poisson
```

```julia
# Gera aleatoriamente contagens de Poisson com média/lambda 6
res = [rand(Poisson(6)) for i in 1:1000]
# Conta cada outcome
contagens = countmap(res)
# Extrai x e y
x = Int.(keys(contagens))
y = values(contagens)./1000
# Plota
bar(x, y, label = "Média = 6")
title!("Distribuição de Poisson")
xticks!(0:15)
```

O histograma acima apresenta uma distribuição de frequência típica de observações. Note-se que a probabilidade maior está concentrada próximo da origem e que a probabilidade de observar grandes valores da variável é bastante pequena. Consequentemente, quando se diz que uma v.a. tem distribuição de Poisson, o que se quer dizer é que a distribuição de frequência de ocorrências para aquela v.a. pode ser razoavelmente aproximada com o uso de uma distribuição de Poisson.

# A fórmula de Poisson

Se uma v.a. é descrita por uma distribuição de Poisson, então a probabilidade de realizar qualquer número dado de ocorrências por unidade de medida é dada pela fórmula

$$P(x) = \frac{e^{-\lambda t}(\lambda t)^x}{x!}$$

onde x é o número de ocorrências; $e$ é a base dos logaritmos naturais; $\lambda$ é a taxa média por unidade e $t$ é o número de unidades. A quantidade $\lambda t$ representa o número médio de ocorrêncais no internvalo $t$. Assim, $\mu = \lambda t$. A fórmula pode ser escrita de forma mais simples substituindo $\lambda t$ por $\mu$:

$$P(x) = \frac{e^{-\mu}(\mu)^x}{x!}$$

#### Exemplo

Um processo mecânico produz tecido para tapetes com uma média de dois defeitos por jarda. Determine a probabilidade de uma jarda quadrada ter exatamente um defeito, admitindo-se que o processo possa ser aproximado por uma distribuição de Poisson.

**Solução**: É dado $\mu = 2$ e $e^{-2} = 0,135$. Assim:

$$P(x = 1) = \frac{e^{-2}(2)^1}{1!} = \frac{0,135(2)}{1} = 0,270$$

#### Exemplo

Suponhamos que os navios cheguem a um porto à razão de $\lambda = 2$ navios/hora, e que essa razão seja bem aproximada por um processo de Poisson. Observando o processo durante um período de meia hora ($t = 1/2$), determine a probabilidade de (a) não chegar nenhum navio, (b) chegarem 3 navios.

**Solução**: Primeiro, determinamos $\mu$:

$$\mu = \lambda t = 2 \left( \frac{1}{2} \right) = 1$$

Pela tabela F, $e^{-1} = 0,368$.

$$(a) \quad P(x = 0) = \frac{e^{-1}(1)^0}{0!} = \frac{e^{-1}(1)}{1} = 0,368$$

$$(b) \quad P(x = 3) = \frac{e^{-1}(1)^3}{3!} = \frac{e^{-1}(1)}{6} = 0,061$$

```julia
# Implementando uma função de probabilidade de poisson
function prob_poisson(x, lambda, t; rounded=true)
    res = ( exp(-(lambda*t)) * (lambda*t)^x) / factorial(x)
    if rounded
        return round(res, digits=4)
    else
        return res
    end
end
```

# A distribuição de Poisson como Aproximação Binomial

Sob certas circunstâncias, a distribuição de Poisson pode ser utilizada para aproximar probabilidades binomiais. A aproximação é mais adequada quando o número de $n$ observações é grande e a probabilidade de sucesso, $p$, está próxima de 0 ou próxima de 1. Convém dispor de um método alternativo para o cálculo de probabilidades binomiais por estas razões:

1. A distribuição binomial descreve adequadamente muitas situações de interesse;
2. A maioria das tabelas está limitada a $n \leq 20$;
3. A fórmula binomial pode exigirt esforço substancial para a obtenção de uma solução exasta.

A vantagem da aproximação reside no fato de que a precisão sofre muito pouco e que o trabalho necessário é consideravelmente menolr. 

#### Exemplo

Determinar a probabilidade de haver 4 peças defeituosas numa amostra de 300, extraída de um grande lote onde há 2% de defeituosas.

**Solução**: Como os valores n = 300 e p = 0,02 estão fora do âmbito de nossas tabelas binomiais, as alternativas seriam: apelar para uma tabela binomial mais extensa, ou calcular

$$P(x = 4) = {300 \choose 4} (0,02)^4 (0,98)^{296}$$

ou utilizar a aproximação de Poisson com $\mu = np$. Assim,

$$\mu = np = 300(0,02) = 6$$

Pela fórmula de Poisson

$$P(x = 4) = \frac{\mu^x \cdot e^{-\mu}}{x!} = \frac{6^4 e^{-6}}{4!} = 0,135$$

Pode-se chegar ao mesmo resultado pela tabela de Poisson: quando $\mu = 6$, pela tabela $P(x = 4) = 0,135$

# Outras distribuições discretas

## A distribuição multinomial

A distribuição multinomial é usada onde há *mais de dois* resultados mutuamente excludentes. Tal como na binomial, exige-se que as provas sejam independentes, com probabilidade constante. Um exemplo da distribuição multinomial é a jogada de um dado que tem seis resultados.

A probabilidade multinomial de que, em $n$ observações, o resultado $E_1$ ocorra $x_1$ vezes, $E_2$ ocorra $x_2$ vezes, ... , e $E_k$ ocorra $x_k$ vezes é dada pela fórmula

$$\frac{n!}{n_1! n_2! \cdots n_k!} ({P_1}^{n1} {P_2}^{n2} \cdots {P_k}^{nk})$$

com

$$n = \sum_{i=1}^{k}n_i$$

e $P_i$ é a probabilidade do evento $E_i$ em cada observação.

#### Exemplo
Um administrador toma conhecimento de que 80% da produção de uma máquina é aceitável, 15% necessita de algum reparo e 5% é imprestável. Numa amostra de $n = 10$ itens, qual é a probabilidade de obter 8 itens bons, 2 que necessitem de reparos, e nenhum imprestável?

$$\frac{10!}{8!2!0!}(0,80)^8(0,15)^2(0,05)^0 = 0,17$$

# A distribuição hipergeométrica

A distribuição hipergeométrica refere-se a situações com dois ou mais resultados em que a probabilidade de sucesso *varia* de uma prova para outra. Por exemplo, suponhamos 10 nomes, um escrito em cada uma de 10 fichas, e colocadas numa urna. Seu nome está numa dessas fichas. Misturam-se as fichas e extrai-se uma. A probabilidade de ela conter o seu nome é $\frac{1}{10}$. Acontece, entretanto, que o nome não é o seu. Extrai-se então uma segunda ficha, das 9 restantes. A probabilidade de ela conter seu nome é agora $\frac{1}{9}$, enquanto que na prova anterior era $\frac{1}{10}$. Logo, a probabilidade desse resultado na segunda extração é *condicional*, ou seja, depende do resultado da primeira extração. Assim, se o primeiro nome extraído tivesse sido o seu, a probabilidade de seu nome na segunda extração seria 0.

Em geral, as probabilidades hipergeométricas podem ser obtidas da fórmula

$$P(x|N, r, n) = \frac{ {N - r \choose n-x} {r \choose x} }{{N \choose n}}$$

onde $N$ é o tamanho da população, $n$ é o tamanho da amostra, $r$ é o número de sucessos na população, e $x$ é o número de sucessos na amostra.

#### Exemplo

Numa caixa com 10 fusíveis, 2 são defeituosos. Extraída uma mostra de 4, qual a probabilidade de (a) nenhum defeituoso, (b) 1 defeituoso, (c) 1 ou menos defeituoso?

**Solução**: Temos $N = 10$, $r = 2$, $n=4$.

$$(a) \quad \quad P(x = 0) = \frac{ {10-2 \choose 4-0} {2 \choose 0} }{{10\choose 4}} = \frac{ {8 \choose 4} {2 \choose 0} }{{10\choose 4}} = 0,333$$

$$(b) \quad \quad P(x = 1) = \frac{ {10-2 \choose 4-1} {2 \choose 1} }{{10\choose 4}} = \frac{ {8 \choose 3} {2 \choose 1} }{{10\choose 4}} = 0,533$$

$$(c) \quad \quad P(x \leq 1) = P(0) + P(1) = 0,333 + 0,533 = 0,866$$

```julia
# Função de prob_multinom para 3 classes
function prob_multinom(ns, ps)
    res = (factorial(sum(ns)) / (factorial(ns[1]) * factorial(ns[2]) * factorial(ns[3]))) * 
        ps[1]^ns[1] * ps[2]^ns[2] * ps[3]^ns[3]
    return res
end
```

```julia
# p. 128 1(b)
prob_multinom([10,0,0], [0.4, 0.3, 0.3])
```

```julia
# p. 128 1(c)
prob_multinom([4,3,3], [0.4, 0.3, 0.3])
```

```julia
# Função de probabilidade hipergeométrica
function prob_hipergeom(x, N, n, r)
    return (binomial(N-r, n-x) * binomial(r, x)) / binomial(N, n)
end
```
